```python
from datascience import *
from datascience import Table
import matplotlib
matplotlib.use('Agg')
from datascience import Table
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

```python
# Change in water is the saturated soil weight subtracted by the measured weight
mini_data = Table().read_table('mini-trial-soil-weight-data.4.csv').relabeled('Date (day of September, 2024)', 'Date')
m_mini_data = mini_data.where('Soil', 'Manteca')
rr_mini_data = mini_data.where('Soil', 'Russell Ranch')
mini_data
```

```python
m_1 = m_mini_data.where('Treatment', 'Control').column('Measured Weight (g)')
m_2 = m_mini_data.where('Treatment', 'Compost').column('Measured Weight (g)')
m_3 = m_mini_data.where('Treatment', 'Compost + Biochar').column('Measured Weight (g)')

fig = plt.figure(figsize =(10, 7))

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Manteca Soil Weight before watering across Treatments')
ax.set_xlabel('Treatments: Control (1), Compost (2), Compost + Biochar (3)')
ax.set_ylabel('Measured Weight (g)')

# Creating plot
bp = ax.boxplot([m_1, m_2, m_3], showmeans=True)

# show plot
plt.show()
```

```python
rr_1 = rr_mini_data.where('Treatment', 'Control').column('Measured Weight (g)')
rr_2 = rr_mini_data.where('Treatment', 'Compost').column('Measured Weight (g)')
rr_3 = rr_mini_data.where('Treatment', 'Compost + Biochar').column('Measured Weight (g)')

fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Russell Ranch Soil Weight before watering across Treatments')
ax.set_xlabel('Treatments: Control (1), Compost (2), Compost + Biochar (3)')
ax.set_ylabel('Measured Weight (g)')

bp = ax.boxplot([rr_1, rr_2, rr_3], showmeans=True)
plt.show()
```

```python
m_1b = m_mini_data.where('Treatment', 'Control').column('Change in water (g)')
m_2b = m_mini_data.where('Treatment', 'Compost').column('Change in water (g)')
m_3b = m_mini_data.where('Treatment', 'Compost + Biochar').column('Change in water (g)')

fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Manteca Soil Change in water across Treatments')
ax.set_xlabel('Treatments: Control (1), Compost (2), Compost + Biochar (3)')
ax.set_ylabel('Change in water (g)')
bp = ax.boxplot([m_1b, m_2b, m_3b], showmeans=True)

# show plot
plt.show()
```
